```python

import pandas as pd

# Load data
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vQkC5sLOdpoyzxkMm3ax22OZIKZ99kUBa8AuiJG2xGSCnwgX28xSkoF6fCoR2WRyE0WTz4m-kQESChv/pub?gid=1808016370&single=true&output=csv'
who_df = pd.read_csv(url)

# Filter dataset for year 2020
df_2020 = who_df[who_df["year"] == 2020].copy()

# Map income group again for the filtered data
income_map = {
    "Low": 0,
    "Lower-middle": 1,
    "Upper-middle": 2,
    "High": 3
}
df_2020["income_group_num"] = df_2020["income group"].map(income_map)



import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Create subplots: 2 rows, 2 columns
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=[
        "Health Exp. (% of GDP) by Income Group",
        "Health Exp. per Capita (USD) by Income Group",
        "Private Health Exp. (USD) by Income Group",
        "Gov. Health Exp. per Capita (USD, 2022 prices) by Income Group"
    ]
)

# 1. Health Exp. (% of GDP)
box1 = px.box(df_2020, x="income group", y="Health Exp. (% of GDP)")
for trace in box1.data:
    fig.add_trace(trace, row=1, col=1)

# 2. Health Exp. per Capita
box2 = px.box(df_2020, x="income group", y="Health Exp. per Capita (USD)")
for trace in box2.data:
    fig.add_trace(trace, row=1, col=2)

# 3. Private Health Exp.
box3 = px.box(df_2020, x="income group", y="Private Health Exp. (USD)")
for trace in box3.data:
    fig.add_trace(trace, row=2, col=1)

# 4. Gov. Health Exp. per Capita
box4 = px.box(df_2020, x="income group", y="Gov. Health Exp. per Capita (USD, 2022 prices)")
for trace in box4.data:
    fig.add_trace(trace, row=2, col=2)

# Update layout
fig.update_layout(
    height=800, width=1000,
    title="Boxplot Subplots of Health Expenditure Variables by Income Group (2020)",
    showlegend=False
)

fig.show()
```

```python
# add visualization code here

app = Dash("Plots and Text")

# Figure 1: Health Expenditure (% of GDP)
fig1 = px.box(
    df_2020,
    x="income group",
    y="Health Exp. (% of GDP)",
    color="income group",
    title="Health Expenditure (% of GDP) by Income Group (2020)"
)

fig1.add_annotation(
    x="High",
    y=14,
    text="Highest median spending",
    showarrow=True,
    arrowhead=2,
    arrowcolor="#b37800",
    font=dict(size=12, color="#b37800"),
    bgcolor="white"
)
fig1.add_annotation(
    x="Low",
    y=5.5,
    text="Lowest overall share of GDP spent on health",
    showarrow=True,
    arrowhead=2,
    arrowcolor="#b37800",
    font=dict(size=12, color="#b37800"),
    bgcolor="white"
)

fig1.update_layout(
    title_x=0.5,
    font=dict(family="Arial", size=13),
    plot_bgcolor="#fafafa",
    paper_bgcolor="#ffffff",
    showlegend=False
)

fig1.update_traces(marker=dict(opacity=0.7))


# Figure 2: Health Expenditure per Capita (USD)
fig2 = px.box(
    df_2020,
    x="income group",
    y="Health Exp. per Capita (USD)",
    color="income group",
    title="Health Expenditure per Capita (USD) by Income Group (2020)"
)

fig2.add_annotation(
    x="High",
    y=11000,
    text="Large gap between groups",
    showarrow=True,
    arrowhead=2,
    arrowcolor="#b37800",
    font=dict(size=12, color="#b37800"),
    bgcolor="white"
)
fig2.add_annotation(
    x="Low",
    y=200,
    text="Under $200 per person in low-income countries",
    showarrow=True,
    arrowhead=2,
    arrowcolor="#b37800",
    font=dict(size=12, color="#b37800"),
    bgcolor="white"
)

fig2.update_layout(
    title_x=0.5,
    font=dict(family="Arial", size=13),
    plot_bgcolor="#fafafa",
    paper_bgcolor="#ffffff",
    showlegend=False
)

fig2.update_traces(marker=dict(opacity=0.7))


# Dash Layout
app.layout = html.Div([
    html.H2("Plots and Text", style={'textAlign': 'center', 'color': '#b37800'}),

    # first visualization
    dcc.Graph(figure=fig1),
    html.P("Takeaway: High-income countries spend a higher percentage of their GDP on healthcare than lower-income groups."),
    html.Br(),

    # second visualization
    dcc.Graph(figure=fig2),
    html.P("Takeaway: Per-capita health spending rises sharply with income, showing large disparities in global healthcare investment.")
])


# DO NOT EDIT BELOW THIS LINE (except to change `jupyter_height`)
if __name__ == '__main__':
    app.run(debug=True, jupyter_mode="inline", jupyter_height=1000)
```
